```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_excel("IT Expenditure dataset.xlsx")

# Convert month abbreviation to datetime
df["Date"] = pd.to_datetime(df["Date"], format="%b")
df["MonthNo"] = df["Date"].dt.month
df["Month"] = df["Date"].dt.strftime("%b")

# Fill missing values
financial_cols = ["Actual", "Forecast", "Plan"]
df[financial_cols] = df[financial_cols].fillna(0)

# Variance metrics
df["Actual vs Plan Variance"] = df["Actual"] - df["Plan"]
df["Actual vs Forecast Variance"] = df["Actual"] - df["Forecast"]

df.head()
```

```python
month_group = df.groupby(["MonthNo","Month"])[
    ["Actual","Plan","Forecast","Actual vs Plan Variance","Actual vs Forecast Variance"]
].sum().reset_index()

month_group = month_group.sort_values("MonthNo")

plt.figure(figsize=(12,5))
plt.bar(month_group["Month"], month_group["Actual vs Plan Variance"])
plt.title("Actual vs Plan Variance by Month (Sorted)")
plt.xlabel("Month")
plt.ylabel("Variance")
plt.axhline(0, linewidth=1)
plt.tight_layout()
plt.show()
```

```python
def plot_top_bottom_variance(df, dim, variance_col, top_n=5):
    grouped = df.groupby(dim)[variance_col].sum().sort_values(ascending=False)

    top = grouped.head(top_n)
    bottom = grouped.tail(top_n)

    combined = pd.concat([top, bottom]).sort_values()

    plt.figure(figsize=(12,5))
    plt.bar(combined.index.astype(str), combined.values)
    plt.title(f"Top & Bottom {top_n} - {variance_col} by {dim}")
    plt.xlabel(dim)
    plt.ylabel(variance_col)
    plt.xticks(rotation=45, ha="right")
    plt.axhline(0, linewidth=1)
    plt.tight_layout()
    plt.show()

dimensions = ["Business Area", "Country", "Cost element name", "IT Area"]

for dim in dimensions:
    print(f"\n--- Visualizing {dim} ---")
    plot_top_bottom_variance(df, dim, "Actual vs Plan Variance")
    plot_top_bottom_variance(df, dim, "Actual vs Forecast Variance")
```
